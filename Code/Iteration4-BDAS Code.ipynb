```python
import findspark
findspark.init('/home/ubuntu/spark-3.2.1-bin-hadoop2.7')
import pyspark
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('basics').getOrCreate()
```

```python
df = spark.read.csv('File_2_ID_2015_Domains_of_deprivation.csv', inferSchema=True, header=True)
df.printSchema()
df.count()
```

```python
for attributes, dtype in df.dtypes:
    print(attributes, dtype)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

pandas_data = df.toPandas()

sns.displot(data=pandas_data, x="Crime Decile", y="Income Decile")
sns.displot(data=pandas_data, x="Crime Decile", y="Employment Decile")
sns.displot(data=pandas_data, x="Income Decile", y="Education Decile")
```
